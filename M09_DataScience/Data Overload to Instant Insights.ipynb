```python
import pandas as pd

# Sample dataset with news articles
data = {'title': ['AI Revolution in Healthcare'],
        'content': ["Artificial Intelligence is transforming healthcare by improving diagnostics, automating administrative tasks, and providing predictive analytics for disease prevention. AI models are being trained on vast medical datasets to assist doctors in making more informed decisions, reducing errors, and optimizing treatment plans. Hospitals worldwide are implementing AI-powered solutions to enhance patient care and streamline operations. However, concerns about data privacy and regulatory compliance remain significant challenges in AI adoption."]}

df = pd.DataFrame(data)
print(df.head())
```

```python
import re

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'\[.*?\]', '', text)  # Remove references
    return text.strip()

df['clean_content'] = df['content'].apply(clean_text)
print(df['clean_content'][0])
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")

summary = summarizer(df['clean_content'][0], max_length=50, min_length=20, do_sample=False)
print("Summarized Text:", summary[0]['summary_text'])
```
