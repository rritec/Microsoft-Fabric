## What is the Resources Section?

The **Resources** section in a Fabric Environment is a lightweight shared file storage area  
that every notebook attached to that environment can read from instantly.

```
Fabric Environment
├── Libraries      ← install Python/R/Java packages   (needs Publish)
├── Spark compute  ← configure executor size/pool     (needs Publish)
└── Resources  ←── shared file storage               (NO Publish needed!)
    ├── config.yml
    ├── lookup.csv
    ├── sales_summary.sql
    ├── schema.json
    └── helpers.py
```

## Key Facts

| Property | Value |
|----------|-------|
| Publish required to take effect | **NO** — changes are real-time |
| Max total storage | **500 MB** |
| Max single file size | **100 MB** |
| Shared across all attached notebooks | **YES** |
| Supports folders | **YES** |
| Access path in notebook | `/env/Resources/` or drag-and-drop generates code |

## Resources vs Lakehouse — When to Use Which

| Store here | When |
|-----------|------|
| **Resources** | Config files, SQL scripts, lookup CSVs, JSON schemas, helper scripts |
| **Lakehouse** | Actual data (Delta tables, large files >100MB, pipeline outputs) |
| **Custom Library** | Python code that needs to be `import`ed as a module across sessions |

## SECTION 3 — Example 1: `config.yml` — Centralised Pipeline Config

Store all your pipeline settings in one YAML file in Resources.  
When you need to change a table name, threshold, or environment flag —  
just edit the file in the Resources panel. No code changes. No republish.

### The `config.yml` file we uploaded:

```yaml
pipeline:
  name: "Sales Pipeline"
  version: "1.0.0"
  environment: "development"

lakehouse:
  name: "sales_lakehouse"
  tables:
    customers: "dim_customers"
    orders: "fact_orders"
    audit: "pipeline_audit_log"

warehouse:
  name: "sales_warehouse"
  schema: "dbo"

data_quality:
  max_null_pct: 5
  max_duplicate_pct: 1
  numeric_cols: ["amount", "quantity"]
  key_cols: ["order_id"]
```

````python
# Cell 2 — Read config.yml from Resources and use it to drive the pipeline

import yaml

CONFIG_PATH = "./env/config.yml"

with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)

print("Config loaded successfully")
print(f"  Pipeline    : {config['pipeline']['name']} v{config['pipeline']['version']}")
print(f"  Environment : {config['pipeline']['environment']}")
print(f"  Lakehouse   : {config['lakehouse']['name']}")
print(f"  Warehouse   : {config['warehouse']['name']}")
print(f"  Key cols    : {config['data_quality']['key_cols']}")
print(f"  Max null %  : {config['data_quality']['max_null_pct']}%")
````

````python
# Cell 3 — Use config values to drive your pipeline dynamically
# All settings come from config.yml — no hardcoded values in notebook code

# Extract values from config
LAKEHOUSE_NAME   = config['lakehouse']['name']
WAREHOUSE_NAME   = config['warehouse']['name']
WAREHOUSE_SCHEMA = config['warehouse']['schema']
TABLE_CUSTOMERS  = config['lakehouse']['tables']['customers']
TABLE_ORDERS     = config['lakehouse']['tables']['orders']
TABLE_AUDIT      = config['lakehouse']['tables']['audit']
KEY_COLS         = config['data_quality']['key_cols']
NUMERIC_COLS     = config['data_quality']['numeric_cols']
MAX_NULL_PCT     = config['data_quality']['max_null_pct']

print("Pipeline settings loaded from config.yml:")
print(f"  Customers table : {TABLE_CUSTOMERS}")
print(f"  Orders table    : {TABLE_ORDERS}")
print(f"  Audit table     : {TABLE_AUDIT}")
print(f"  Key cols        : {KEY_COLS}")
print(f"  Numeric cols    : {NUMERIC_COLS}")
print(f"  Max null %      : {MAX_NULL_PCT}%")

# Now use these variables throughout the notebook
# e.g. spark.sql(f'SELECT * FROM {TABLE_ORDERS}')
# When the table name changes — just update config.yml, no notebook changes needed
````

## SECTION 4 — Example 2: `lookup.csv` — Reference / Enrichment Table

Small lookup tables used for enriching data, joining on codes, or filtering —  
stored in Resources so every notebook gets the same reference data.

### The `lookup.csv` file we uploaded:

```
category,category_code,department,tax_rate,is_active
Electronics,ELEC,Technology,0.08,true
Clothing,CLTH,Fashion,0.05,true
Furniture,FURN,Home,0.07,true
...
```

````python
# Cell 4 — Read lookup.csv from Resources and join it with Lakehouse data

import pandas as pd

LOOKUP_PATH = "./env/lookup.csv"

# Read CSV from Resources using pandas
lookup_pd = pd.read_csv(LOOKUP_PATH)
print(f"Lookup table loaded: {len(lookup_pd)} categories")
print(lookup_pd.to_string(index=False))
````

````python
# Cell 5 — Convert lookup to Spark DataFrame and join with fact_orders

# Convert pandas → Spark for joining at scale
lookup_spark = spark.createDataFrame(lookup_pd)

# Read orders from Lakehouse
orders_df = spark.sql(f"SELECT * FROM {TABLE_ORDERS}")

# Join orders with lookup to enrich with department, tax_rate
enriched_df = orders_df.join(
    lookup_spark.select("category", "category_code", "department", "tax_rate"),
    on="category",
    how="left"
)

# Calculate tax amount using enriched tax_rate from Resources lookup
from pyspark.sql import functions as F

enriched_df = enriched_df.withColumn(
    "tax_amount",
    F.round(F.col("total_amount") * F.col("tax_rate"), 2)
)

print(f"Orders enriched with lookup data: {enriched_df.count()} rows")
enriched_df.select(
    "order_id", "category", "category_code",
    "department", "total_amount", "tax_rate", "tax_amount"
).show(8, truncate=False)
````

## SECTION 5 — Example 3: `sales_summary.sql` — Shared SQL Scripts

Store commonly used SQL queries in Resources so any analyst or engineer  
on the team can run them without copying SQL into every notebook.

### The `sales_summary.sql` file we uploaded:

```sql
SELECT
    category,
    COUNT(order_id)             AS total_orders,
    SUM(total_amount)           AS total_revenue,
    AVG(total_amount)           AS avg_order_value,
    MIN(total_amount)           AS min_order_value,
    MAX(total_amount)           AS max_order_value,
    COUNT(DISTINCT customer_id) AS unique_customers
FROM fact_orders
WHERE status = 'Completed'
GROUP BY category
ORDER BY total_revenue DESC
```

````python
# Cell 6 — Read SQL from Resources and execute it

SQL_PATH = "./env/sales_summary.sql"

with open(SQL_PATH, "r") as f:
    sql_query = f.read().strip()

print("SQL loaded from Resources:")
print("-" * 50)
print(sql_query)
print("-" * 50)

# Execute the shared SQL query
summary_df = spark.sql(sql_query)

print(f"\nQuery result: {summary_df.count()} categories")
summary_df.show(truncate=False)
````

````python
# Cell 7 — Pattern: load multiple SQL files from a queries/ subfolder
# This shows how to organise SQL scripts into folders in Resources

# Assume you have Resources/queries/ folder with multiple .sql files
# e.g. Resources/queries/daily_revenue.sql
#      Resources/queries/top_customers.sql
#      Resources/queries/refund_report.sql

import os

def run_sql_from_resources(filename: str, base_path: str = "/env/Resources") -> None:
    """
    Reads a .sql file from Resources and executes it with spark.sql().

    Args:
        filename  : filename including any subfolder e.g. 'queries/daily.sql'
        base_path : Resources root path (default: /env/Resources)
    """
    full_path = os.path.join(base_path, filename)
    with open(full_path, "r") as f:
        query = f.read().strip()
    print(f"Executing: {filename}")
    return spark.sql(query)

# Execute from root Resources folder
result = run_sql_from_resources("sales_summary.sql")
result.show(5)
````

## SECTION 6 — Example 4: `schema.json` — Data Contract Validation

Store your table schema contracts as JSON files in Resources.  
Every notebook can validate incoming data against the agreed contract —  
catching schema drift before it breaks downstream reports.

### The `schema.json` file we uploaded:

```json
{
  "table": "fact_orders",
  "version": "1.0.0",
  "columns": [
    {"name": "order_id",    "type": "IntegerType", "nullable": false},
    {"name": "customer_id", "type": "StringType",  "nullable": false},
    {"name": "amount",      "type": "DoubleType",  "nullable": false},
    ...
  ]
}
```

````python
# Cell 8 — Load schema.json from Resources and validate a DataFrame against it

import json

SCHEMA_PATH = "/env/Resources/schema.json"

with open(SCHEMA_PATH, "r") as f:
    schema_contract = json.load(f)

print(f"Schema contract loaded: {schema_contract['table']} v{schema_contract['version']}")
print(f"Defined columns: {len(schema_contract['columns'])}")
print("\nContract columns:")
for col in schema_contract['columns']:
    nullable = "nullable" if col['nullable'] else "NOT NULL"
    print(f"  {col['name']:<20} {col['type']:<15} {nullable}")
````

````python
# Cell 9 — Validate fact_orders DataFrame against the schema contract

orders_df = spark.sql(f"SELECT * FROM {TABLE_ORDERS}")

# Get actual schema from the DataFrame
actual_schema = {f.name: str(f.dataType) for f in orders_df.schema.fields}

print(f"{'Column':<22} {'Contract Type':<18} {'Actual Type':<18} {'Match'}")
print("-" * 68)

all_passed = True
for col_def in schema_contract['columns']:
    col_name     = col_def['name']
    expected     = col_def['type']
    actual       = actual_schema.get(col_name, 'MISSING')
    match        = "PASS" if expected == actual else "FAIL"
    if match == "FAIL":
        all_passed = False
    print(f"  {col_name:<20} {expected:<18} {actual:<18} {match}")

print("-" * 68)
if all_passed:
    print("  Schema validation PASSED — DataFrame matches contract")
else:
    print("  Schema validation FAILED — check FAIL rows above")
````

## SECTION 7 — Example 5: `helpers.py` — Shared Utility Script

Store a Python utility script in Resources and load it via `importlib`  
in any notebook — **without re-publishing the Environment**.

### Difference from Custom Library `.py`

| | Custom Library `.py` | Resources `helpers.py` |
|---|---|---|
| How to update | Re-upload + Publish (5-10 min) | Edit file in Resources — instant |
| Import style | `from helpers import func` | `importlib.util.spec_from_file_location` |
| Best for | Stable shared code | Frequently changing utilities |

### The `helpers.py` file we uploaded contains:
- `load_config(path)` — loads a YAML config file
- `load_schema(path)` — loads a JSON schema contract
- `load_sql(path)` — loads a SQL script string
- `validate_schema(df, schema)` — validates DataFrame against JSON contract
- `print_config_summary(config)` — pretty prints config summary

````python
# Cell 10 — Import helpers.py from Resources using importlib
# importlib lets you load a .py file from any path — no package structure needed

import importlib.util
import sys

HELPERS_PATH = "./env/helpers.py"

spec   = importlib.util.spec_from_file_location("helpers", HELPERS_PATH)
helpers = importlib.util.module_from_spec(spec)
spec.loader.exec_module(helpers)

print("helpers.py loaded from Resources")
print(f"Available functions: {[f for f in dir(helpers) if not f.startswith('_')]}")
````

````python
# Cell 11 — Use helpers.py functions to load all Resources files in one go

# Use helpers to load config
config    = helpers.load_config("./env/config.yml")
helpers.print_config_summary(config)

# Use helpers to load SQL
sql_query = helpers.load_sql("./env/sales_summary.sql")
print(f"\nSQL loaded ({len(sql_query)} chars):")
print(sql_query[:120] + "...")

# Use helpers to load and validate schema
schema_contract = helpers.load_schema("./env/schema.json")
orders_df       = spark.sql(f"SELECT * FROM {TABLE_ORDERS}")

print(f"\nRunning schema validation using helpers.validate_schema()...")
validation = helpers.validate_schema(orders_df, schema_contract)

print(f"Total checks : {validation['total']}")
print(f"Passed       : {validation['passed']}")
print(f"Failed       : {validation['failed']}")
print(f"All passed   : {validation['all_passed']}")

# Show any failed columns
if not validation['all_passed']:
    print("\nFailed columns:")
    for col, r in validation['checks'].items():
        if not r['passed']:
            print(f"  {col}: expected {r['expected']}, got {r['actual']}")
````

## SECTION 8 — Putting It All Together

A complete pipeline that reads **all configuration from Resources** —  
zero hardcoded values in the notebook code.

````python
# Cell 12 — Full pipeline driven entirely by Resources files
#
# Resources files used:
#   config.yml         → table names, thresholds, settings
#   schema.json        → validate incoming data contract
#   sales_summary.sql  → run summary report
#   lookup.csv         → enrich orders with department and tax rate
#   helpers.py         → utility functions

import importlib.util, yaml, json, os
from pyspark.sql import functions as F

R = "./env/"   # Resources root shorthand

# ── Step 1: Load all config from Resources ───────────────────────────────────
spec    = importlib.util.spec_from_file_location("helpers", f"{R}/helpers.py")
helpers = importlib.util.module_from_spec(spec)
spec.loader.exec_module(helpers)

config          = helpers.load_config(f"{R}/config.yml")
schema_contract = helpers.load_schema(f"{R}/schema.json")
sql_query       = helpers.load_sql(f"{R}/sales_summary.sql")

ORDERS_TABLE = config['lakehouse']['tables']['orders']
MAX_NULL_PCT = config['data_quality']['max_null_pct']

print("[1] All config loaded from Resources")

# ── Step 2: Read data from Lakehouse ─────────────────────────────────────────
orders_df = spark.sql(f"SELECT * FROM {ORDERS_TABLE}")
print(f"[2] Lakehouse read: {orders_df.count():,} rows from {ORDERS_TABLE}")

# ── Step 3: Validate schema against contract from Resources ──────────────────
validation = helpers.validate_schema(orders_df, schema_contract)
status     = "PASS" if validation['all_passed'] else "FAIL"
print(f"[3] Schema validation: {status} ({validation['passed']}/{validation['total']} checks passed)")

# ── Step 4: Enrich with lookup CSV from Resources ────────────────────────────
import pandas as pd
lookup_spark = spark.createDataFrame(pd.read_csv(f"{R}/lookup.csv"))

enriched_df = orders_df.join(
    lookup_spark.select("category", "department", "tax_rate"),
    on="category", how="left"
).withColumn("tax_amount", F.round(F.col("total_amount") * F.col("tax_rate"), 2))

print(f"[4] Lookup enrichment complete: added department + tax_rate + tax_amount")

# ── Step 5: Run summary SQL from Resources ───────────────────────────────────
summary_df = spark.sql(sql_query)
print(f"[5] SQL summary executed: {summary_df.count()} category rows")

# ── Step 6: Check null threshold from config ─────────────────────────────────
null_counts = {c: orders_df.filter(F.col(c).isNull()).count() for c in orders_df.columns}
total_rows  = orders_df.count()
null_issues = {
    c: round(n / total_rows * 100, 1)
    for c, n in null_counts.items()
    if n > 0
}

print(f"[6] Null check (threshold from config: {MAX_NULL_PCT}%):")
if null_issues:
    for col, pct in null_issues.items():
        flag = "PASS" if pct <= MAX_NULL_PCT else "FAIL"
        print(f"    {col}: {pct}%  [{flag}]")
else:
    print("    No nulls found — PASS")

print("\n[DONE] Pipeline completed — driven entirely by Resources files")
summary_df.show()
````

## SECTION 9 — Managing Resources via Notebook Code

You can also create, read, update, and delete files in Resources  
programmatically from within a notebook.

````python
# Cell 13 — Create, read, update, delete files in Resources from notebook

import os, json

R = "./env/"

# ── CREATE — write a new file to Resources ───────────────────────────────────
run_log = {
    "pipeline"    : "Sales Pipeline",
    "last_run"    : "2024-01-15 14:30:00",
    "rows_loaded" : 2000,
    "status"      : "success"
}
with open(f"{R}/run_log.json", "w") as f:
    json.dump(run_log, f, indent=2)
print("Created: run_log.json")

# ── READ — read the file back ─────────────────────────────────────────────────
with open(f"{R}/run_log.json", "r") as f:
    loaded = json.load(f)
print(f"Read back: {loaded}")

# ── UPDATE — overwrite with new content ──────────────────────────────────────
loaded['last_run'] = "2024-01-16 09:00:00"
loaded['rows_loaded'] = 2150
with open(f"{R}/run_log.json", "w") as f:
    json.dump(loaded, f, indent=2)
print("Updated: run_log.json")

# ── CREATE FOLDER ────────────────────────────────────────────────────────────
os.makedirs(f"{R}/queries", exist_ok=True)
print("Created folder: Resources/queries/")

# ── LIST all files recursively ────────────────────────────────────────────────
print("\nAll files in Resources:")
for root, dirs, files in os.walk(R):
    level = root.replace(R, '').count(os.sep)
    indent = '  ' * level
    folder = os.path.basename(root)
    if level > 0:
        print(f"{indent}{folder}/")
    for file in sorted(files):
        size = os.path.getsize(os.path.join(root, file))
        print(f"{indent}  {file}  ({size} bytes)")

# ── DELETE ───────────────────────────────────────────────────────────────────
os.remove(f"{R}/run_log.json")
print("\nDeleted: run_log.json")
````

## SECTION 10 — Troubleshooting

| Problem | Cause | Fix |
|---------|-------|-----|
| `FileNotFoundError: /env/Resources/config.yml` | File not uploaded yet OR wrong filename | Go to Environment → Resources → verify file is there |
| File updated in UI but notebook reads old content | Notebook session caches file handle | Re-open the file in a new cell — no restart needed |
| `FileNotFoundError` for a file in a subfolder | Path wrong | Use `/env/Resources/subfolder/filename.ext` |
| `500 MB limit exceeded` when uploading | Total Resources size reached 500 MB | Delete old files or move large data to Lakehouse |
| `100 MB` single file too large | File exceeds per-file limit | Split file or store in Lakehouse instead |
| `helpers.py` changes not reflected | Old module cached in Python memory | Re-run the `importlib` cells to reload the module |
