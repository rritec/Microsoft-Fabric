# Define Parameters

```python
bronzeFilePath="" #"Files/bronze/emp.csv"
lakeHouse=""#"FAB_DH_LAKEHOUSE"
silverSchema=""#"silver"
silverDeltaTable=""#"emp"
```

# parameters validation

```python
assert bronzeFilePath != "", "bronzeFilePath parameter missing"
assert lakeHouse != "", "lakeHouse parameter missing"
assert silverSchema != "", "silverSchema parameter missing"
assert silverDeltaTable != "", "silverDeltaTable parameter missing"
```

# process

```python
# Read emp.csv from Lakehouse Files (Bronze)
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(bronzeFilePath)

#df.show()
```

```python
from pyspark.sql.functions import col, coalesce, lit

df_processed = df.withColumn(
    "totalsal",
    col("sal") + coalesce(col("comm"), lit(0))
)

#df_processed.show()
```

```python
df_processed.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(f"{lakeHouse}.{silverSchema}.{silverDeltaTable}")
```

```python
df = spark.sql(f"SELECT * FROM {lakeHouse}.{silverSchema}.{silverDeltaTable} LIMIT 1000")
#df.show(2)
```
